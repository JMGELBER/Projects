
## Import the data

```r
data = read.csv(file = "data.csv")
```

## Creating Time Series for Speed and Power

```r
speed = ts(data$Speed_ms)
power = ts(data$Power_Watt)
```

## Plot for wind speed over time

```r
plot(speed, xlab="Time", ylab="Speed(m/s)", main="Wind Speed over Time",
     lwd=1.5, cex=1.2, cex.axis=1.2, cex.lab=1.2, pch=19, col="blue", xaxt="none", panel.first=grid())
# for visualizing partition of each day(referring to row indices in excel file)
axis(1, at=c(1,85,229,373), labels=c("01/09", "01/10", "01/11", "01/12"), cex.axis=0.8)
abline(v=c(1,85,229,373), col="black", lty=2, lwd=0.5)
legend(legend="Daily Peak(Time/Value)", "topleft", pch=19, cex=0.7, col="red")

a = which.max(speed[1:84])
b = max(speed[1:84])
points(a, b, pch=19, cex=1, col="red")
text(a, b+0.3, labels="20:40/2.163", col="orange", font=3, cex=0.7)

c = 84 + which.max(speed[85:228])
d = max(speed[85:228])
points(c, d, pch=19, cex=1, col="red")
text(c, d+0.3, labels="15:00/2.951", col="orange", font=3, cex=0.7)

e = 228 + which.max(speed[229:372])
f = max(speed[229:372])
points(e, f, pch=19, cex=1, col="red")
text(e-30, f, labels="23:40/4.343", col="orange", font=3, cex=0.7)

g = 372 + which.max(speed[373:387])
h = max(speed[373:387])
points(g, h, pch=19, cex=1, col="red")
text(e-20, f+0.3, labels="00:50/4.742", col="orange", font=3, cex=0.7)
```

## Plot for wind power over time

```r
plot(power, xlab="Time", ylab="Power(W)", main="Wind Power over Time",
     lwd=1.5, cex=1.2, cex.axis=1.2, cex.lab=1.2, pch=19, col="blue", xaxt="none", panel.first=grid())
# for visualizing partition of each day
axis(1, at=c(1,85,229,373), labels=c("01/09", "01/10", "01/11", "01/12"), cex.axis=0.8)
abline(v=c(1,85,229,373), col="black", lty=2, lwd=0.5)
legend(legend="Daily Peak(Time/Value)", "topleft", pch=19, cex=0.7, col="red")

# daily peak time on plot
a2 = which.max(power[1:84])
b2 = max(power[1:84])
points(a2, b2, pch=19, cex=1, col="red")
text(a2, b2+3, labels="22:50/28.891", col="orange", font=3, cex=0.7)

c2 = 84 + which.max(power[85:228])
d2 = max(power[85:228])
points(c2, d2, pch=19, cex=1, col="red")
text(c2, d2+3, labels="15:00/40.214", col="orange", font=3, cex=0.7)

e2 = 228 + which.max(power[229:372])
f2 = max(power[229:372])
points(e2, f2, pch=19, cex=1, col="red")
text(e2-30, f2, labels="23:40/62.921", col="orange", font=3, cex=0.7)

g2 = 372 + which.max(power[373:387])
h2 = max(power[373:387])
points(g2, h2, pch=19, cex=1, col="red")
text(e2-20, h2, labels="01:10/80.522", col="orange", font=3, cex=0.7)
```

## Plot for speed and power over time

```r
# binded plots
par(mfrow=c(2,1), mar=c(1,4,2,1.5))

## plot for speed
plot(speed, ylab="Speed(m/s)", lwd=1.5, cex=1, cex.axis=1, cex.lab=1, xaxt="none", panel.first=grid())
axis(1, at=c(1,85,229,373), labels=c("01/09", "01/10", "01/11", "01/12"), cex.axis=0.8)
abline(v=c(a,c,e,g), col="red", lty=2)

## plot for power
plot(power, ylab="Power(W)", lwd=1.5, cex=1, cex.axis=1, cex.lab=1, col="blue", xaxt="none", panel.first=grid())
abline(v=c(a2,c2,e2,g2), col="red", lty=2)
```

## Plot speed on x-axis versus power on y-axis

```r
par(mfrow=c(1,1), mar=c(5.1,4.1,4.1,2.1)) # for original figure
plot(speed, power, xlab="Speed(m/s)", ylab="Power(W)", main="Wind Power over Wind Speed",
     lwd=1.5, cex=1, cex.axis=1, cex.lab=1, pch=21, col="blue", panel.first=grid())
abline(v=1.494, col="red", lty=2)
text(1,60, labels="1.494m/s ->", col="red", cex=1)
```

The estimated
cut-off wind speed value is (about) 1.494 m/s. Above 1.494 m/s, the wind
speed seems to play a role in generating power with linear
proportionality.
